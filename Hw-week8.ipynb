# Exploratory Analysis

```python
# to convert the data in the same data type
pd.options.mode.copy_on_write = True 
data["Season_A_Production"] = data["Season_A_Production"].astype(str)
data["Total_Production"] = data["Total_Production"].astype(str)

data["Season_A_Production"]= data["Season_A_Production"].str.replace(",","").astype(float)
data
data["Total_Production"] = data["Total_Production"].str.replace(",","").astype(float)
data
```

```python
# calculation of mean, median and standard deviation
data.describe()
```

```python
# to sort the data and the top maize production District is Nyagatare  

sorted_df = data.sort_values('Total_Production' , ascending=False)
sorted_df
```

# Analyze distribution of maize production. Create histograms or boxplots to visualize the spread of production values

```python
# analysis of the distribution of maize using Histogram
import matplotlib as plt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(14, 7))

plt.subplot(1, 3, 1)
sns.histplot(data['Season_A_Production'], kde=True)
plt.title('Season A Production')

plt.subplot(1, 3, 2)
sns.histplot(data['Season_B_Production'], kde=True)
plt.title('Season B Production')

plt.subplot(1, 3, 3)
sns.histplot(data['Total_Production'], kde=True)
plt.title('Total Production')

plt.tight_layout()
plt.show()

```

```python
# Create Boxplots
plt.figure(figsize=(14, 7))

plt.subplot(1, 3, 1)
sns.boxplot(y=data['Season_A_Production'])
plt.title('Season A Production')

plt.subplot(1, 3, 2)
sns.boxplot(y=data['Season_B_Production'])
plt.title('Season B Production')

plt.subplot(1, 3, 3)
sns.boxplot(y=data['Total_Production'])
plt.title('Total Production')

plt.tight_layout()
plt.show()
```

# Visualize maize production on a map of Rwanda (using libraries like geopandas)

```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

rwanda_map = gpd.read_file('C:\\Users\\educa\\Desktop\\Hw-week8\\rwa_district')
rwanda_maize = rwanda_map.merge(data)


fig, ax = plt.subplots(1, 1, figsize=(10, 10))
rwanda_maize.plot(column='Total_Production', ax=ax, legend=True,
                  legend_kwds={'label': "Maize Production (tons)",
                               'orientation': "horizontal"},
                  cmap='YlGn')
for idx, row in rwanda_maize.iterrows():
    plt.annotate(text=row['District'],  # Replace 'district_name' with the actual column name for district names
                 xy=(row.geometry.centroid.x, row.geometry.centroid.y),
                 horizontalalignment='center', fontsize=8, color='black')

# Set title and show the plot
ax.set_title('Maize Production in Rwanda by District')
plt.show()
```
