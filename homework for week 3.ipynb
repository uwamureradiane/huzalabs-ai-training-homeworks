# Question 3( Borrowing system)

```python
class Book:
    def __init__(self, title, author, isbn, availability=True):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.availability = True  # private attribute

    def display_info(self):
        print("Title:", self.title)
        print("Author:", self.author)
        print("ISBN:", self.isbn)
        print("Availability:", "Available" if self.availability else "Not Available")

    # Getter method for availability
    def get_availability(self):
        return self.availability

    # Setter method for availability
    def set_availability(self, availability):
        self.availability = availability


class Library:
    def __init__(self):
        self.book = []

    def add_book(self, book):
        self.book.append(book)
        print("Book added successfully.")

    def remove_book(self, book):
        if book in self.book:
            self.book.remove(book)
            print("Book removed successfully.")
        else:
            print("Book not found in the library.")

    def list_books(self):
        if self.book:
            print("List of Books:")
            for book in self.book:
                print("-" * 20)
                book.display_info()
        else:
            print("No books in the library.")

    def search_book(self, search_name):
        found_book = []
        for book in self.book:
            if search_name.lower() in book.title.lower() or search_name.lower() in book.author.lower():
                found_book.append(book)
        if found_book:
            print("Found Book:")
            for found_book in found_book:
                print("-" * 20)
                found_book.display_info()
        else:
            print(f"No books found matching '{search_name}'.")

    def check_out_book(self, title):
        for book in self.book:
            if book.title.lower() == title.lower():
                if book.get_availability():
                    book.set_availability(False)
                    print(f"'{title}' checked out successfully.")
                else:
                    print(f"'{title}' is not available for checkout.")
                return
        print(f"'{title}' not found in the library.")

    def return_book(self, title):
        for book in self.book:
            if book.title.lower() == title.lower():
                if not book.get_availability():
                    book.set_availability(True)
                    print(f"'{title}' returned successfully.")
                else:
                    print(f"'{title}' is already available in the library.")
                return
        print(f"'{title}' not found in the library.")

# Example usage
if __name__ == "__main__":
    # Create some book objects
    book1 = Book("Python Programming", "John Smith", "978-0134853987")
    book2 = Book("Java Basics", "Alice Johnson", "978-0321773702")
    book3 = Book("Data Science Handbook", "Emily Brown", "978-1491912058")

    # Create a library
    library = Library()

    # Add books to the library
    library.add_book(book1)
    library.add_book(book2)
    library.add_book(book3)

    # List all books in the library
    library.list_books()

    # Check out a book
    print("\nChecking out 'Python Programming'...")
    library.check_out_book("Python Programming")

    # Attempt to check out an unavailable book
    print("\nChecking out 'Python Programming' again...")
    library.check_out_book("Python Programming")

    # Return a book
    print("\nReturning 'Python Programming'...")
    library.return_book("Python Programming")

    # List all books after returning a book
    print("\nList of Books after return:")
    library.list_books()
```
